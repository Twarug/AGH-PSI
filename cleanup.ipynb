```python
import pandas as pd
import ast
import json
import os 
```

```python
df = pd.read_csv("dataset/movies.csv", usecols=['genres','original_language', 'original_title', 'production_companies', 'production_countries', 'release_date', 'runtime', 'vote_average', 'vote_count'])

print(df.shape)
df = df.dropna()
print(df.shape)

df
```

```python
def extract_field(json, field):
    try:
        list_string = ast.literal_eval(json)

        if not isinstance(list_string, list):
            return ""
        
        names=[item.get(field) for item in list_string if isinstance(item, dict) and item.get(field)]

        return ", ".join(names)
    except (SyntaxError, ValueError):
        return ""
    
```

```python
df['genres'] = df['genres'].apply(lambda x: extract_field(x, 'name'))
df['production_companies'] = df['production_companies'].apply(lambda x: extract_field(x, 'name'))
df['production_countries'] = df['production_countries'].apply(lambda x: extract_field(x, 'name'))
df
```

```python
output_directory = "dataset"
output_filename = "cleaned_movies_data.csv"
output_path = os.path.join(output_directory, output_filename)

try:
    df.to_csv(output_path, index=False, encoding='utf-8')
    print(f"\nDataFrame zostaÅ‚ pomyÅ›lnie zapisany do pliku: {output_path}")
except Exception as e:
    print(f"\nWystÄ…piÅ‚ bÅ‚Ä…d podczas zapisywania pliku CSV: {e}")
```
